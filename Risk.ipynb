```python
## !/usr/bin/python3
# -*- coding: utf-8 -*-
 
#Milosza

from tkinter import *
import numpy as np
import random as rand
from tkinter import scrolledtext

#Frame
root = Tk();        
root.title('Risk 1.3')
Frame(root, width=450, height=600).pack() #rozmiar
Label(root, text='-=[ Risk ]=-').place(x=180, y=5)
Label(root, text='Attack', fg='red').place(x=90, y=20)    
Label(root, text='Defend', fg='blue').place(x=290, y=20)

attack = IntVar();
defend = IntVar();
attack.set(5)
defend.set(5)

#Displays
aDisplay = Entry(root, textvariable = attack, relief=RIDGE, bd = 2, width=10,    insertwidth = 1, font = 40);
aDisplay.place(x=50, y=40);
dDisplay = Entry(root, textvariable = defend, relief=RIDGE, bd = 2, width=10,    insertwidth = 1, font = 40);
dDisplay.place(x=250, y=40);

#Order functions
def clear():
    aDisplay.delete(0,END);
    dDisplay.delete(0,END);
    return;

def update_attack(v):
    attack.set(v)

def update_defend(u):
    defend.set(u)

#Main function
def clear_log():
    txt.delete(1.0,END)
    
    attack = attack.get()
    defend = defend.get()
    print('attack ' + str(attack))
    print('defend ' + str(defend))
    print('times  ' + str(times))
    tekst = 'attack ' + str(attack) + ' defend ' + str(defend) + ' times  ' + str(times)
    logger(tekst,"INFO")
    

def risk(attack,defend,times):
    
    attack = attack.get()
    defend = defend.get()
    stop = 1
    
    print('===============================') 
    logger('# status             # attack # defend #',"INFO")
    logger('#--------------------------------------#',"INFO")
    logger('# WAR HAS BEGUN !!!  ' +
           '# ' + str(attack) + (7-len(str(attack))) * ' ' +
           '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"INFO")
    
    while attack > 0 and defend > 0 and stop > 0:
        
        if times == 0:
            stop = 0
        
        logger('#--------------------------------------#',"INFO")
        defend_dice_list = [0,0]
        attack_dice_list = [0,0,0]         
        
        if defend > 1: 
            defend_dice_num = 2 
        else: 
            defend_dice_num = 1
        if attack > 2: 
            attack_dice_num = 3 
        else: 
            if attack > 1: 
                attack_dice_num = 2 
            else: 
                attack_dice_num = 1
        print('attack dice: ' + str(attack_dice_num))
        print('defend dice: ' + str(defend_dice_num))
        log = 'attack dice: ' + str(attack_dice_num)
        log = 'defend dice: ' + str(defend_dice_num)

        defend_dice_list[0] = rand.randint(1, 6)
        if defend_dice_num > 1:
            defend_dice_list[1] = rand.randint(1, 6)
        
        list.sort(defend_dice_list, reverse = True)
        print('defend_dice_list: ' + str(defend_dice_list))
        
        defend_dice_list

        attack_dice_list[0] = rand.randint(1, 6)
        if attack_dice_num > 1:
            attack_dice_list[1] = rand.randint(1, 6)
        if attack_dice_num > 2:
            attack_dice_list[2] = rand.randint(1, 6)
        
        list.sort(attack_dice_list, reverse = True)
        print('attack_dice_list: ' + str(attack_dice_list))
        
        attack_dice_list
        
        #iterate and compare
        #print results
        
        #new soldier number
        
        if attack_dice_list[0] > defend_dice_list[0]:
            defend = defend - 1
            logger('# ' + str(attack_dice_list[0]) + ' ---> ' + str(defend_dice_list[0]) + '           # ' 
                        + str(attack) + (7-len(str(attack))) * ' ' +
                   '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"ATTACK")
        else:
            attack = attack - 1
            logger('# ' + str(attack_dice_list[0]) + ' <--- ' + str(defend_dice_list[0]) + '           # ' 
                        + str(attack) + (7-len(str(attack))) * ' ' +
                   '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"DEFEND")
        if attack_dice_num > 1 and defend_dice_num > 1:
            if attack_dice_list[1] > defend_dice_list[1]:
                defend = defend - 1
                logger('# ' + str(attack_dice_list[1]) + ' ---> ' + str(defend_dice_list[1]) + '           # ' 
                            + str(attack) + (7-len(str(attack))) * ' ' +
                       '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"ATTACK")                
            else:
                attack = attack - 1
                logger('# ' + str(attack_dice_list[1]) + ' <--- ' + str(defend_dice_list[1]) + '           # ' 
                            + str(attack) + (7-len(str(attack))) * ' ' +
                       '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"DEFEND")                
        #end while loop
        
    logger('#======================================#',"INFO")
    
    if attack > defend and defend == 0:
        logger('# defend lost...     # ' 
                    + str(attack) + (7-len(str(attack))) * ' ' +
               '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"INFO")
    else:
        if attack < defend and attack == 0:
            logger('# attack lost...     # ' 
                        + str(attack) + (7-len(str(attack))) * ' ' +
                   '# ' + str(defend) + (7-len(str(defend))) * ' ' + '#',"INFO")
        
    update_attack(attack)
    update_defend(defend)

    return

#Buttons
battle    = Button(root, text='Battle',    width=8, height=5, bg='Red', 
                   command = lambda: risk(attack,defend,0)).place(x=264, y=482);

anihilate = Button(root, text='Total War', width=8, height=5, bg='Red', 
                   command = lambda: risk(attack,defend,1)).place(x=334, y=482);

clear     = Button(root, text='Clear',     width=10, height=5, 
                   command = clear, bg='Orange').place(x=180, y=482);


######################
 
tekst = StringVar();
    
txt = scrolledtext.ScrolledText(root,width=40,height=24)
txt.place(x=50, y=80)
txt.configure(font='TkFixedFont')
txt.tag_config('INFO', foreground='black')
txt.tag_config('DEFEND', foreground='blue')
txt.tag_config('ATTACK', foreground='red')


def logger(tekst,mod): 
    #tekst = won.get()
    #mod = mod.get()
    txt.insert(INSERT, tekst + '\n',mod)

log = Button(root, text="Clear Log", width=8, height=5, bg='Blue', 
             command = clear_log).place(x=40, y=482)
 

######################


root.configure(background = 'black');
root.mainloop();
```
